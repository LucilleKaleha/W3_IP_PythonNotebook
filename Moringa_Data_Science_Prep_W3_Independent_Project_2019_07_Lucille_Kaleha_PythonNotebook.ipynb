```
b = pd.read_csv('MoringaSchool_CB_CDR_20120507.csv', )
b.columns = ['PRODUCT',	'VALUE', 'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
b.head()
```

```
c = pd.read_csv('MoringaSchool_CB_CDR_20120508.csv', )
c.columns = ['PRODUCT',	'VALUE', 'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
c.head()
```

# Data Preparation

```
# merging the three datasets in order to get a better comparison
e = pd.concat([b, c, d], ignore_index= True)
e.head()
```

```
# dropping the unrequired columns
e.drop('COUNTRY_A', axis=1, inplace=True)
e.drop('COUNTRY_B', axis=1, inplace=True)
e.drop('CELL_ON_SITE', axis=1, inplace=True)
e.head()
```

```
# new dataset with split date and time columns 
new = e["DATE_TIME"].str.split(" ", n = 1, expand = True) 
  
# making separate date column from new data frame 
e["DATE"]= new[0] 
  
# making separate time column from new dataset
e["TIME"]= new[1] 
  
# Dropping old DATE_TIME columns 
e.drop(columns =["DATE_TIME"], inplace = True) 
e.head()
```
